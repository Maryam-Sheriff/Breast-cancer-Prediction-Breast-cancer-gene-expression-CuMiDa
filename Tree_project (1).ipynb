```python
import numpy as np
```

```python
import pandas as pd
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("Breast_GSE45827.csv")
```

```python
df.dropna()
```

```python
Droped=df.drop(columns = ['samples'])
```

```python
print(Droped)
```

```python
Droped.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
Droped["type"] = ord_enc.fit_transform(Droped[["type"]])
Droped["type"]
```

```python
Droped.head()
```

```python
y=Droped['type']
```

```python
X=Droped.drop(columns = ['type'])
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=324)
```

```python
max_leaf_nodes = range(2,100,2) 
testing_accuracy = []
training_accuracy = []
score = 0
```

```python
for i in max_leaf_nodes: 
    dt = DecisionTreeClassifier(max_leaf_nodes=i, random_state=342)
    dt.fit(X_train,y_train)
    y_pred = dt.predict(X_train)
    training_accuracy.append(accuracy_score(y_train, y_pred)) 
    y_predict_test = dt.predict(X_test)
    acc_score = accuracy_score(y_test,y_predict_test)
    testing_accuracy.append(acc_score) 
    if score < acc_score: 
        score = acc_score 
        best_depth = i
```

```python
print(best_depth)
```

```python
print("accuracy for best depth:" , score*100)
```
